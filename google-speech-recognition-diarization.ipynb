```python
from google.cloud import speech_v1p1beta1
from google.cloud.speech_v1 import types
types.RecognitionConfig.AudioEncoding.LINEAR16


```

```python
def transcribe_hindi_audio_with_diarization(audio_file_path):
    client = speech_v1p1beta1.SpeechClient()

    # The language code for Hindi
    language_code = "hi-IN"

    # Configure the diarization settings
    enable_diarization = True
    diarization_speaker_count = 2

    config = {
        "enable_speaker_diarization": enable_diarization,
        "diarization_speaker_count": diarization_speaker_count,
        "language_code": language_code,
    }

    # Read the audio file
    with open(audio_file_path, "rb") as audio_file:
        audio = speech_v1p1beta1.RecognitionAudio(content=audio_file.read())

    # Perform the speech recognition with diarization
    response = client.recognize(config=config, audio=audio)

    # Process the response
    for result in response.results:
        alternative = result.alternatives[0]
        print("Transcript: {}".format(alternative.transcript))
        print("Speaker Tag: {}".format(alternative.words[-1].speaker_tag))


```

```python
# Provide the path to your Hindi audio file
audio_file_path = "/kaggle/input/audio1/test (2).wav"

```

```python
# Call the function to transcribe the audio with diarization
transcribe_hindi_audio_with_diarization(audio_file_path)
```
